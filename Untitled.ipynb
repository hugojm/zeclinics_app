```python
import os
from PIL import Image, ImageTk
import os
import torch
from torchvision import transforms
from torchvision.models import *
import torchvision
from matplotlib import pyplot as plt
from matplotlib.figure import Figure
import numpy as np
```

```python
device = 'cpu'
model = torch.load('static/weight/weights.pt' , map_location = torch.device('cpu'))
img = Image.open('/Volumes/GoogleDrive/.shortcut-targets-by-id/19L0VOL58iClS-3O8Wl4vWZHavaxx7oh6/UPC/BAT1/20190903_1694_CS_R2/Well_right_F01/20190903_1951_R2_1694_R2_head_W_right_F01_1_4.png')
data_transforms = transforms.Compose([transforms.ToTensor()])
data_img = data_transforms(img)
inputs = data_img.unsqueeze(0).to(device)
outputs = model(inputs)
cmaps = ['Purples', 'Oranges','Blues', 'Greens', 'Blues', 'Reds']
plt.imshow(torch.split(data_img.unsqueeze(0),1,0)[0].squeeze().permute(1, 2, 0))
for i in range(6):
    mask = torch.split(outputs['out'].cpu(),1,0)[0].squeeze()[i].detach().numpy()
    mask = np.ma.masked_where(mask < 0.05, mask)
    plt.imshow(mask, cmaps[i], alpha = 0.7)
    
```
